```python
import pandas as pd
import numpy as np
```

```python
 X = pd.read_csv('C:\\Users\\suleiman\\Documents\\scikit\\csv_files\\X 15.csv', header = None )
```

```python
X, y = X.iloc[:,1:].values, X.iloc[:,0].values
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
sel = SelectKBest(chi2, k='all')
```

```python
sel.fit(X, y)
```

```python
sel.scores_
```

```python
score = sel.scores_.round(3)
```

```python
import matplotlib.pyplot as plt
```

```python
indices= (['Circularity','High Luminance Scale','FAST Corners','Aspect Ratio','Solidity','Average Pixel Intensity','Average Luminance',
           'Entropy','Color Skewness','Shape Skewness','Boundary Skewness','Color Kurtosis','Shape Kurtosis','Boundary Kurtosis','Variance'])
```

```python
plt.figure(figsize=(10,7))
plt.title("Feature importances", fontsize=20)
plt.bar(range(X.shape[1]), score,color="r",  width=.4 , align="center", )
#indices= (['1','HLS','FC','AR','SOL','API','ALI','ENT','ClSkw','SpSkw','BdSkw','ClKur','SpKur','BdKur','VAR'])
plt.xticks(range(X.shape[1]), indices, rotation='vertical', fontsize=20)
#plt.xticks(np.arange(len(a)), np.arange(1, len(a)+1), fontsize=15)
plt.yticks(fontsize=15)
plt.xlim([-1, X.shape[1]])
plt.show() 
```
